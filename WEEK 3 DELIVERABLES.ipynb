```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
pd.options.display.float_format = '{:.0f}'.format
```

```python
retail_sales = pd.read_csv('Europe_GDP.csv') 
data = retail_sales.copy()
data.head()
```

```python
data['Year'] = data['Year'].astype(int)
```

```python
austrias_gdp= data[['Year', 'Austria']]
austrias_gdp= austrias_gdp.set_index('Year')
austrias_gdp.head()
```

# LINE PLOT

```python
plt.figure(figsize=(10,8))

plt.plot(austrias_gdp.index, austrias_gdp.Austria, marker='o', linestyle= '-', color= 'orange')

plt.title('Austrias GDP (1960 - 2023)')
plt.xlabel('Year')
plt.ylabel('GDP')

plt.ylim(austrias_gdp['Austria'].min(), austrias_gdp['Austria'].max())

plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
austrias_gdp.isna().sum()
```

# SCATTER PLOT

```python
plt.figure(figsize=(10,7))

plt.scatter(data.Germany, data.Austria, marker='o', color= 'orange')

plt.title('Relationship between Germany and Austria GDP')
plt.xlabel('Germany')
plt.ylabel('Austria')

plt.show()
```

# HISTOGRAM

```python
for col in data.columns:
    sb.histplot(data[col])
    plt.title(col.upper())
    plt.ylabel('Count')
    plt.xlabel('Frequency count')
    plt.show()
```

# BOXPLOT

```python
for col in data.select_dtypes(include='number').columns:
    sb.boxplot(data[col])
    plt.title(col)
    plt.ylabel('Values')
    plt.show()
```
